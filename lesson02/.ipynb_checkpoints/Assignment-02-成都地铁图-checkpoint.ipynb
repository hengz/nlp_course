### 1.2 爬取本地宝地铁数据

````python
import requests
import re

base_url = "http://cd.bendibao.com/ditie"
line1_suffix = "/xl_13.shtml"
response = requests.get(base_url+line1_suffix)
````

````python
raw_page = response.text

patten_text_lines = r'<a href="/ditie(.+)" target="_blank">地铁'
patten_lines = re.compile(patten_text_lines)
line_suffixes = patten_lines.findall(raw_page)
metro_line_urls = [base_url + x for x in line_suffixes]
metro_line_urls.insert(0, base_url+line1_suffix)
metro_line_urls
````

````python
line_station_sets = {}
for metro_line_url in metro_line_urls:
    response = requests.get(metro_line_url)
    raw_page = response.text
    
    patten_text_metro = r'<strong>(\w+)</strong>'
    patten_text_station = r'<td><a href="/ditie/\w+.shtml" target="_blank">(\w+)</a></td>'
    
    patten_metro = re.compile(patten_text_metro)
    patten_station = re.compile(patten_text_station)
    
    key = patten_metro.findall(raw_page)
    value = patten_station.findall(raw_page)
    line_station_sets[key[0]] = value
````

使用geopy包，估计国内地址不全，这个方法查不到地址

````python
from geopy.exc import GeocoderTimedOut
from geopy.geocoders import Nominatim
import geopy

def do_geocode(address): 
    try:
        geolocator = Nominatim(user_agent="metro_location", timeout=3)
        return geolocator.geocode(address)
    except GeocoderTimedOut:
        return do_geocode(address)

location = do_geocode("成都地铁韦家碾站")

print(location.address)
print((location.latitude, location.longitude))
print(location.raw)
````
